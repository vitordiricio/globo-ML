```python
import pandas as pd
import re
```

```python
linear_semana = pd.read_csv('tv_linear_semanal.csv', sep=';')
```

```python
linear_semana
```

```python
linear_semana.columns = linear_semana.columns.str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('[', '').str.replace(']', '')
linear_semana = linear_semana[linear_semana['faixas_horárias'] == 'Dia Inteiro'].reset_index(drop=True)
linear_semana['emissoras'] = linear_semana['emissoras'].str.replace(''', '').str.replace(''', '').str.replace(' - ', ' ').str.replace('é', 'e').str.replace(' ', '_').str.replace('‘', '').str.replace('’', '')
linear_semana['emissoras'] = linear_semana['emissoras'].str.replace('TV_BAND', 'TVBAND').str.replace('NI_�NIC_at�_Mar19�', 'NINICATEHMAR19').str.replace('NI_TOTAL', 'NITOTAL')
linear_semana = linear_semana[['datas','emissoras', 'rat%', 'shr%', 'tvr%', 'avrch%_wavg', 'cov%', 'tvr#']]
```

```python
linear_semana
```

```python
# Linha 1: Substituir vírgulas por pontos em todas as colunas numéricas
linear_semana.loc[:, linear_semana.columns.difference(['datas', 'emissoras'])] = linear_semana.loc[:, linear_semana.columns.difference(['datas', 'emissoras'])].replace(',', '.', regex=True)

# Linha 2: Converter todas as colunas numéricas para float
linear_semana.loc[:, linear_semana.columns.difference(['datas', 'emissoras'])] = linear_semana.loc[:, linear_semana.columns.difference(['datas', 'emissoras'])].astype(float)
```

```python
# Função para extrair a data de início da semana
def extrair_data_inicio(texto_data):
    # Usar regex para extrair a data de início entre parênteses
    padrao = r'\((\d{2}/\d{2}/\d{4})'
    match = re.search(padrao, texto_data)
    if match:
        return match.group(1)
    return None

# Aplicar a função na coluna 'datas' para obter a data de início
linear_semana['data_inicio'] = linear_semana['datas'].apply(extrair_data_inicio)

# Converter para datetime usando o formato brasileiro (dia/mês/ano)
linear_semana['data_inicio'] = pd.to_datetime(linear_semana['data_inicio'], format='%d/%m/%Y')

# Se preferir, pode criar uma coluna para armazenar a semana do ano
linear_semana['semana_ano'] = linear_semana['data_inicio'].dt.isocalendar().week
linear_semana['ano'] = linear_semana['data_inicio'].dt.isocalendar().year

# Opcional: criar uma chave de semana no formato "YYYY-WW"
linear_semana['semana_chave'] = linear_semana['data_inicio'].dt.strftime('%Y-W') + linear_semana['semana_ano'].astype(str).str.zfill(2)

# Se quiser preservar também a informação do intervalo original, 
# pode criar colunas para data de início e fim
def extrair_data_fim(texto_data):
    padrao = r'- (\d{2}/\d{2}/\d{4})'
    match = re.search(padrao, texto_data)
    if match:
        return match.group(1)
    return None

linear_semana['data_fim'] = linear_semana['datas'].apply(extrair_data_fim)
linear_semana['data_fim'] = pd.to_datetime(linear_semana['data_fim'], format='%d/%m/%Y')

# Se precisar, pode também extrair as informações de semana/ano da string original
def extrair_semana_ano(texto_data):
    padrao = r'Week (\d+)/(\d+)'
    match = re.search(padrao, texto_data)
    if match:
        return (int(match.group(1)), int(match.group(2)))
    return (None, None)

linear_semana['semana_original'], linear_semana['ano_abreviado'] = zip(*linear_semana['datas'].apply(extrair_semana_ano))
```
